```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7,8]
y=[3,4,5,6,7,3,2]
c=["y","g","b","r","y","g","b"]

plt.scatter(x,y,color=c,marker="*",s=400,edgecolor="r",linewidth=2)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7,8]
y=[3,4,5,6,7,3,2]
colors=[10,43,45,34,23,30,25]

plt.scatter(x,y,c=colors,marker="*",s=400,edgecolor="r",linewidth=2,cmap="viridis")
plt.colorbar()
plt.show()
```
