```python
#stack plot or Area plot 

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
plt.stackplot(x,area1)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[3,4,5,6,7]
area3=[3,4,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3, labels=l)
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[3,4,5,6,7]
area3=[3,4,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3, labels=l,baseline="zero")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.show()
```
