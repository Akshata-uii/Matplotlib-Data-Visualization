```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[3,4,5,6,7,8,6]
z=[3,4,2,5,6,7,5]
c=["r","g","m","y","g","b",'g']
plt.xlabel("numbers",fontsize=15,color="r")
plt.ylabel("% numbers",fontsize=15,color="b")
plt.title("Bar_chart",fontsize=15)
plt.bar(x,y,color=c,label="WScubetech",alpha=0.6,edgecolor="r",linestyle=":")
plt.bar(x,z,label="WScubetech2",edgecolor="r",linestyle=":")

plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7]
y=[3,4,5,6,7,8,6]
z=[3,4,2,5,6,7,5]
c=["r","g","m","y","g","b",'g']

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("numbers",fontsize=15,color="r")
plt.ylabel("% numbers",fontsize=15,color="b")
plt.title("Bar_chart",fontsize=15)

plt.bar(p,y,width,label="WScubetech")
plt.bar(p1,z,width,label="WScubetech2")

plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7]
y=[3,4,5,6,7,8,6]
z=[3,4,2,5,6,7,5]
c=["r","g","m","y","g","b",'g']

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("numbers",fontsize=15,color="r")
plt.ylabel("% numbers",fontsize=15,color="b")
plt.title("Bar_chart",fontsize=15)

plt.bar(p,y,width,label="WScubetech")
plt.bar(p1,z,width,label="WScubetech2")
plt.xticks(p+width/2,x,rotation=20)
plt.legend(loc=2)
plt.show()
```
