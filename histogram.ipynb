```python
#by using axvline we will able to show which data is needful and which data is not 

import matplotlib.pyplot as plt
import numpy as np
import random

no=[51 ,24 ,56, 68 ,12, 50, 32, 50, 13, 57, 36 ,35 ,21 ,23 ,51 ,38, 40 ,28, 47 ,39,66, 10, 53, 45,
 18 ,21 ,44, 65, 56, 63 ,22, 65, 19 ,6,3, 48, 66, 44, 47, 48 ,52, 28 ,18 ,58 ,27 ,39, 64, 14, 23,
 28 ,33]

l=[10,20,30,40,50,60]

#convert to log

plt.hist(no, color="b",edgecolor='r' ,bins=l,rwidth=0.8)
plt.axvline(45, color="g",label="line")
plt.legend()
plt.show()
```
