```python
#stem plot 

import matplotlib.pyplot as plt 
x=[3,4,5,6,6,7]
y=[3,4,5,6,7,8]
plt.stem(x,y)
plt.show()
```

```python

# lineformat
import matplotlib.pyplot as plt 
x=[3,4,5,6,6,7]
y=[3,4,5,6,7,8]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python

import matplotlib.pyplot as plt 
x=[3,4,5,6,6,7]
y=[3,4,5,6,7,8]
plt.stem(x,y,linefmt="-",markerfmt="*")
plt.show()
```

```python

import matplotlib.pyplot as plt 
x=[3,4,5,6,6,7]
y=[3,4,5,6,7,8]
plt.stem(x,y,linefmt="-",markerfmt="*",basefmt="r")
plt.show()
```

```python

# use_line_collection is used to change the colors of lines
import matplotlib.pyplot as plt 
x=[3,4,5,6,6,7]
y=[3,4,5,6,7,8]
plt.stem(x,y,linefmt="-",markerfmt="*",basefmt="r",use_line_collection=False)
plt.show()
```

```python
#orientation 
import matplotlib.pyplot as plt 
x=[3,4,5,6,6,7]
y=[3,4,5,6,7,8]
plt.stem(x,y,linefmt="-",markerfmt="*",basefmt="r",use_line_collection=False,orientation="horizontal")
plt.show()
```
